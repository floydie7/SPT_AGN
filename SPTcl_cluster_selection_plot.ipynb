```python
"""
SPTcl_cluster_selection_plot.ipynb
Author: Benjamin Floyd

Creates a mass-redshift plot of the full SPT-SZ + SPTpol 100d surveys with our SPTcl sample highlighted.
"""

import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
# Read in the original catalogs
bocquet = Table.read('Data_Repository/Catalogs/SPT/SPT_catalogs/2500d_cluster_sample_Bocquet18.fits')
huang = Table.read('Data_Repository/Catalogs/SPT/SPT_catalogs/sptpol100d_catalog_huang19.fits')

# Read in our SPTcl-IRAGN catalog
sptcl = Table.read('Data_Repository/Project_Data/SPT-IRAGN/Output/SPTcl_IRAGN.fits')
```

```python
# Make the plot
fig, ax = plt.subplots()
ax.scatter(bocquet['REDSHIFT'], bocquet['M500'], marker='.', c='tab:gray', label=r'SPT-SZ 2500 deg$^2$')
ax.scatter(huang['redshift'], huang['M500'], marker='^', c='tab:gray', label='SPTpol 100d')
ax.scatter(sptcl['REDSHIFT'], sptcl['M500']/1e14, marker='o', c='tab:blue', label='This work')
ax.legend()
ax.set(xlabel='Redshift', ylabel=r'M$_{500c}$ [10$^{14} M_\odot h^{-1}_{70}$]', xlim=[0, 1.75], yscale='log')
plt.show()
```
